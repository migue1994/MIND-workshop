# Cálculo de estadísticas básicas

```python
df.describe()
```

## Estadisticas por agrupación de datos

## Bluethoot

```python
dfBlue=df.groupby('blue')
```

```python
dfBlue.describe()
```

## Dual sim

```python
dfDualSim = df.groupby('dual_sim')
```

```python
dfDualSim.describe()
```

## Camara fontal en megapixeles

```python
dfFrontCamera = df.groupby('fc')
```

```python
dfFrontCamera.describe()
```

## Disponibilidad 4G

```python
df4G = df.groupby('four_g')
```

```python
df4G.describe()
```

## Number of cores of processor

```python
dCores = df.groupby('n_cores')
dCores.describe()
```

## Primary Camera mega pixels

```python
dfprimaryC = df.groupby('pc')
dfprimaryC.describe()
```

## Has 3G or not

```python
df3G = df.groupby('three_g')
df3G.describe()
```

## Has touch screen or not

```python
dfTouch = df.groupby('touch_screen')
```

```python
dfTouch.describe()
```

## Has wifi or not

```python
dfWifi = df.groupby('wifi')
dfWifi.describe()
```

## Price vs ram

```python
sns.set(style="whitegrid")
plt.figure()
ax = sns.boxplot(x='price_range', y='ram', data=df, orient="v")
```
